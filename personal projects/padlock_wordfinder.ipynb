```python
# consolidated keypad code

import json
from urllib.request import urlopen

dial_characters = ["DRHMTGLSBP", "AOHURLNIEY", "AOTCULERNS", "EKTLADYBN "]

def list_maker(input):
    fulllist = []
    for elem in input:
        newlist = []
        for letter in elem:
            newlist.append(letter)
        fulllist.append(newlist)
    return fulllist

diallist = list_maker(dial_characters)
#diallist = [['A','B'], ['A','B'], ['A','B'], ['A', ' ']]
print(diallist)

iter = 0
master_wordlist = []

while iter < len(diallist[0]):
    for w in diallist[0]:
        for x in diallist[1]:
            for y in diallist[2]:
                for z in diallist[3]:
                    if z == ' ':
                        master_wordlist.append(w+x+y)
                    else: 
                        master_wordlist.append(w+x+y+z)
        iter+= 1
#print(master_wordlist)
print(len(master_wordlist))


f = open('dictionary.json')
data = json.load(f)
websters_dictionary = data
f.close()
#verify file was imported correctly:
print(len(websters_dictionary))
#output: 102217

dictlist = websters_dictionary.keys()
dictlist = [x.lower() for x in dictlist]
dictset = set(websters_dictionary.keys())
print(len(dictset))
#print(dictlist)

keypad = [x.lower() for x in master_wordlist]
keypad = set(keypad)
#print(keypad)

commonwords = keypad.intersection(dictset)
commonwords = sorted(list(commonwords))

commonwordscapital = []
for i in commonwords:
    i = i.capitalize()
    commonwordscapital.append(i)

commonwords = commonwordscapital

print(commonwords)

len(commonwords)
```
