```python
from matplotlib import pyplot as plt

def loop():
    newlist = []
    x = 100
    while x > 4:
        if (x % 2) == 0:
            x = x / 2
            newlist.append(x)
        else:
            x = (x*3)+1
            newlist.append(x)
    else:
        if x == 4:
            x = 2
            newlist.append(x)
            x = 1
            newlist.append(x)
        else:
            x = 1
            newlist.append(x)
    return newlist

listtoplot = loop()
print(listtoplot)
plt.plot(listtoplot)
plt.show
```
