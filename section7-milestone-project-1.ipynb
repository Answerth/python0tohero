```python
# validation of input

def user_choice():
    choice = "wrong"
    acceptable_range = range(0,10)
    within_range = False

    while choice.isdigit() == False or within_range == False:
        choice = input(print("Please enter a number (1-10): "))

        if choice.isdigit() == False:
            print("Sorry that is not a digit!")
        
        if choice.isdigit() == True:  # if choice.isdigit() - implies "if true"
            if int(choice) in acceptable_range:
                within_range = True
            else: 
                print("Not in range")
                within_range = False


    return int(choice)

user_choice()
```

```python
def display_game(game_list):
    print("Here is the current game list: ")
    print(game_list)

def position_choice():
    choice = 'wrong'
    choice_range = ['0', '1', '2']
    while choice not in choice_range:

        choice = input(print("Pick a position (0,1,2): "))

        if choice not in choice_range:
            print("Sorry, invalid choice")

    return int(choice) 

def replacement_choice(game_list, position):

    user_placement = input(print("Type a string to place at position"))

    game_list[position] = user_placement

    return game_list

def game_on_choice():
    choice = 'wrong'
    game_modes = ['Y', 'N']
    while choice not in game_modes:

        choice = input(print("Keep playing? ")).upper()

        if choice not in game_modes:
            print("Sorry, invalid choice")

    if choice.upper() == 'Y':
        return True
    else:
        return False


game_on = True
game_list = [0,1,2]

while game_on:
    display_game(game_list)
    position = position_choice()
    game_list = replacement_choice(game_list, position)
    display_game(game_list)
    game_on = game_on_choice()



display_game(game_list)
position_choice()
replacement_choice(game_list, 1)
game_on_choice()
```
