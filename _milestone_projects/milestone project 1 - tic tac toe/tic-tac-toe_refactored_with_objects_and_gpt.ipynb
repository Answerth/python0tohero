```python
fakeboard = [['X','O','X'], 
            ['O','X','O'], 
            ['X','O','X']]

def draw_board(board):
    # Define the horizontal separator used between rows
    horizontal_separator = "-".join(["+"] + ["---"] * len(board[0])) + "+"
    
    # Draw the board
    board_rows = []
    for row in board:
        # Draw the cells in the current row
        cell_strs = []
        for cell in row:
            if cell is None:
                cell_strs.append("   ")
            elif cell == "X":
                cell_strs.append(" X ")
            elif cell == "O":
                cell_strs.append(" O ")
        
        # Concatenate the cell strings into a row string
        row_str = "|".join([""] + cell_strs + [""]) 
        
        # Add the row string to the list of board rows
        board_rows.append(row_str)
    
    # Concatenate the board rows into the final board string
    board_str = "\n".join([horizontal_separator] + board_rows + [horizontal_separator])
    
    # Print the board string
    print(board_str)

value = draw_board(fakeboard)
print(value)
```
