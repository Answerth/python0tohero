```python
class Board:
    def __init__(self):
        self.board = [[' ',' ',' '], [' ',' ',' '], [' ',' ',' ']]
        
    def print_board(self):
        for i in range(len(self.board)):
            print(" " + self.board[i][0] + " | " + self.board[i][1] + " | " + self.board[i][2] + " ")
            if i != len(self.board)-1:
                print("-----------")
            else:
                break
                
    def update_board(self, row, col, symbol):
        self.board[row][col] = symbol


class Player:
    def __init__(self, name, symbol):
        self.name = name
        self.symbol = symbol
        
    def get_move(self, board):
        valid_move = False 
        count = 0 
        while not valid_move:
            move = input(f"{self.name}, enter your move (cell number): ")
            try: 
                userinput = int(move)
                if userinput < 1 or userinput > 9: 
                    print("Invalid move: cell number must be between 1 and 9.")
                else: 
                    row = (userinput - 1) // 3
                    col = (userinput - 1) % 3
                    if board.board[row][col] != ' ': 
                        print("invalid move")
                    else: 
                        valid_move = True
            except ValueError: 
                print("invalid. enter number 1-9")
# instead of counting to 9 and brute forcing, i should check winner, and if not winner, check if all values != ' ' through list comprehension
# get move could validate board first, then validate valid input, and perhaps call a win/draw from checkwinner        
            count += 1
            if count == 9: 
                print("no more valid moves")
                break
        return (row, col)

    def check_win(self, board):
        # check rows
        for i in range(3):
            if board.board[i][0] == board.board[i][1] == board.board[i][2] == self.symbol:
                return True
        # check columns
        for i in range(3):
            if board.board[0][i] == board.board[1][i] == board.board[2][i] == self.symbol:
                return True
        # check diagonals
        if board.board[0][0] == board.board[1][1] == board.board[2][2] == self.symbol:
            return True
        if board.board[0][2] == board.board[1][1] == board.board[2][0] == self.symbol:
            return True
        return False


def play_game(player1, player2, board):
    current_player = player1
    while True:
        board.print_board() 
        print("__________")
        row, col = current_player.get_move(board)
        board.update_board(row, col, current_player.symbol)
        if current_player.check_win(board):
            board.print_board() 
            print(f"{current_player.name} wins!")
            break
        elif all([all([cell != ' ' for cell in row]) for row in board.board]):
            board.print_board() 
            print("Draw!")
            break
        current_player = player2 if current_player == player1 else player1


board = Board()
def getplayername():
    player = input("what is your name?")
    return player

player1_name = getplayername()
player1 = Player(player1_name, "X")
print(f"{player1.name}, you've been assigned the symbol {player1.symbol}")

player2_name = getplayername()
player2 = Player(player2_name, "O")
print(f"{player2.name}, you've been assigned the symbol {player2.symbol}")

play_game(player1, player2, board)
```
